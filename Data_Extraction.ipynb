```python
#First, imports
import time 
import requests
import json
import csv
import pprint
import pandas as pd
from Config import apikey
```

```python
#Start by accessing api
base_url=f"https://api.polygon.io/v2/aggs/ticker/"
#Beer=BUDweiser and coors(TAP), Soda=COKE and PEPsi, Coffee=dutchBROS and StarBUX 
tickers=["BUD","TAP","COKE","PEP","BROS","SBUX"]
data_list=[]
```

```python
def iterate_months():
    start_month = 11
    for year in range(2021, 2024):
        for month in range(start_month, 12):
            yield f"{year}-{month:02d}-01"
        start_month = 1
```

```python
#Responses in following string
# C- Close price                # O- Open price  
# H- High price                 # T- Unix time
# L- Low price                  # V- Volume
# N- Number of transactions     # VW- Volume weighted
```

```python
#Gather data and append to a list
for tick in tickers:
    
    #Makes the print a little prettier
    print("-------------------------")
    print(f"Now Finding: {tick}")
    print("-------------------------")
    
    #Iterate, find and append data
    for month in iterate_months():
        
        #Shows the date time
        print("-------------------------")
        print(f"Result Date: {month}")
        print("-------------------------")
        
        full_url=base_url+f"{tick}/range/12/month/{month}/{month}?adjusted=true&sort=asc&limit=120&apiKey={apikey}"
        data = requests.get(full_url)
        data_json = data.json()
        formatted_data = pprint.pformat(data_json)
        print(formatted_data)
        data_list.append(data_json)
        
        #limited to 5 pulls per miniute, timer allows for the sixth call
        time.sleep(13)
        
        #Stop on todays date
        if month == "2023-10-01":
            break 
    
#Quality of life to say when its done
print("-------------------------")
print(f"Done!")
print("-------------------------")
```

```python
#Create dataframe
df = pd.DataFrame(data_list)
df.head()
```

```python
#Clean the dataframe
cleandf = df.loc[:, ['ticker', 'results']]
cleandf
```

```python
#Parse the results data
df_new = cleandf.explode('results')
def break_up_dictionary(row):
    results = row['results'][0]
    for key, value in results.items():
        row[key] = value
    return row

df_new = cleandf.apply(break_up_dictionary, axis=1)
df_new=df_new
```

```python
#Save data to csv
df.to_csv('data.csv', index=True)
```
