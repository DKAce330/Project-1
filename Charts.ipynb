```python
import csv
import pprint
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
from Config import apikey
import seaborn as sns
#Import df from csv
path=("data.csv")
df=pd.read_csv(path)
def fix_unix_time(unix_time_values):
    scaled_time_values = (unix_time_values - min(unix_time_values)) / (max(unix_time_values) - min(unix_time_values))
    return scaled_time_values

# Create a list of Unix time values
t = df['t']

# Convert and scale the Unix time values
st = fix_unix_time(t)

# Add the scaled time values to the DataFrame
df['st'] = st

# Print the DataFrame
df

df
```

```python
# Filter the DataFrame by ticker
bud_df = df[df['ticker'] == 'BUD']
coors_df= df[df['ticker'] == 'TAP']
coke_df= df[df['ticker'] == 'KO']
pepsi_df= df[df['ticker'] == 'PEP']
star_df= df[df['ticker'] == 'SBUX']
dutch_df= df[df['ticker'] == 'BROS']
```

```python
#Line graphs

x_values = bud_df['st']
y1_values = bud_df['c']
y2_values = coors_df['c']
y3_values = coke_df['c']
y4_values = pepsi_df['c']
y5_values = star_df['c']
y6_values = dutch_df['c']

plt.plot(x_values, y1_values, label='Budweiser')
plt.plot(x_values, y2_values, label='Coors')
plt.plot(x_values, y3_values, label='Coke')
plt.plot(x_values, y4_values, label='Pepsi')
plt.plot(x_values, y5_values, label='Starbucks')
plt.plot(x_values, y6_values, label='Dutch')

plt.xlabel('Time')
plt.ylabel('Closing Price')
plt.title('Beverage Stocks')
plt.legend()
plt.plot
plt.savefig('Graphs/Complete.png')
```
